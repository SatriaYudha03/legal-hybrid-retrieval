# 03 — Hasil Evaluasi

Membandingkan performa **BM25**, **IndoSBERT**, dan **Hybrid (RRF)** memakai metrik Recall@5, Recall@10, MRR, dan NDCG@10.
Cocok untuk slide *Hasil & Pembahasan*.

**Prasyarat:** sudah menjalankan `python scripts/04_evaluate.py` (yang menghasilkan `results/metrics.json`).

```python
import sys, json
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

import pandas as pd
import matplotlib.pyplot as plt
from src.config import load_config, resolve_path

cfg = load_config()
metrics_path = resolve_path(cfg['paths']['results_dir']) / 'metrics.json'
summary = json.loads(metrics_path.read_text(encoding='utf-8'))

df = pd.DataFrame(summary).T  # baris = model, kolom = metrik
df = df.rename(index={'bm25': 'BM25', 'indosbert': 'IndoSBERT', 'hybrid': 'Hybrid (RRF)'})
df.round(4)
```

## Grafik perbandingan antar-model

```python
ax = df.plot(kind='bar', figsize=(9, 5), colormap='viridis')
ax.set_title('Perbandingan Performa Retrieval Dokumen Hukum')
ax.set_ylabel('Skor'); ax.set_xlabel('Model')
ax.legend(title='Metrik', bbox_to_anchor=(1.0, 1.0))
plt.xticks(rotation=0); plt.tight_layout(); plt.show()
```
